```python
import subprocess
from collections import Counter
from datetime import datetime
from pathlib import Path
```

```python

REPO = Path("../resources/youtube-dl")
```

```python

def git(cmd):
    return subprocess.check_output(["git", *cmd], cwd=REPO).decode().strip()
```

```python
# 1. Ile commitów ma projekt?
total_commits = int(git(["rev-list", "--count", "HEAD"]))
print("Commits:", total_commits)
```

```python

# 2. Kto ma najwięcej commitów?
shortlog = git(["shortlog", "-sn", "HEAD"]).splitlines()
top_line = shortlog[0].strip()
top_count = int(top_line.split()[0])
top_author = " ".join(top_line.split()[1:])
percent = round(top_count / total_commits * 100, 2)

print("Top author:", top_author)
print("Commits:", top_count)
print("Percent:", percent, "%")
```

```python

# 3. Najbardziej aktywny rok/miesiąc
dates = git(["log", "--date=format:%Y-%m", "--format=%ad"]).splitlines()
month_counter = Counter(dates)
most_active_month, most_active_count = month_counter.most_common(1)[0]

print("Most active month:", most_active_month, "commits:", most_active_count)
```

```python

# 4. Najczęściej zmieniane pliki
files = git(["log", "--format=format:", "--name-only"]).splitlines()
files = [f for f in files if f.strip()]
file_counter = Counter(files)

print("Top 20 changed files:")
for name, count in file_counter.most_common(20):
    print(count, name)
```

```python

# 5. Czy projekt jest aktywnie rozwijany?
last_commit_date = git(["log", "-1", "--format=%ad", "--date=iso"])
last_commit = datetime.fromisoformat(last_commit_date)

days_since = (datetime.now(last_commit.tzinfo) - last_commit).days

print("Last commit:", last_commit_date)
print("Days since last commit:", days_since)

if days_since < 90:
    print("Project is likely still active.")
else:
    print("Project is likely inactive.")
```
